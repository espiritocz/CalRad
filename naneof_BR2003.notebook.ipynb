```python
####################################################################################################
# naneof_BR2003.py 
# Karl Lapo July/2015
# Originally written by Brian Henn in MATLAB (doi: 10.1175/JHM-D-12-027.1)
####################################################################################################
# Function for iteratively calculating eofs on discontinous data 
####################################################################################################
#   Function to compute EOFs from data with irregular NaNs.
# Computes EOFs of ROW DIMENSION in d matrix are in columns 
# Variances are in # of total variance ( sum(d(:).^2)/N, not /M/N!)
#
# Follows Becken and Rixen, JTECH 2003 http://dx.doi.org/10.1175/1520-0426(2003)020<1839:ECADFF>2.0.CO;2 
# As written by Brian Henn in MATLAB (doi: 10.1175/JHM-D-12-027.1)
#
# SYNTAX:
#   [B,vars,amps,Nopt] = naneof_br3(data_in)
#
# INPUT:        
#       data_in = MxN numpy array ----- should have overall nanmean value of 0
#
# OUTPUT:
#       B       = MxM numpy array is the PC time series (unitless?), EOF numbers are in columns
#       vars    = Mx1 numpy array, shows variance described by each EOF 
#       amps    = NXM numpy array, spatial weights (EOFs)
#       Nopt    = scalar, the number of optimal EOFs generating the least validation error
#
# Data Reconstruction (this gives the zero-mean filled data matrix):
#   reconstructed = B(:,1:Nopt)*amps(:,1:Nopt)'

## Import statements
# netcdf/numpy/xray
import numpy as np
from datetime import datetime, timedelta

# OS interaction
import sys, pickle, os
```

```python
###### Beckers and Rixen 2003 example
# import subplots function for plotting
import seaborn as sns
import matplotlib
from matplotlib.pyplot import subplots
import matplotlib.pyplot as plt
from matplotlib import cm
import random
# must insert this statement to render the plots within the notebook
# this is specific to the ipython notebook
%matplotlib inline

#Generate data
m=50
n=100
frac_gaps = 0.5 # the fraction of data with NaNs
NSratio = 0.1 # the Noise to Signal ratio for adding noise to data
 
x = (np.arange(m)*2*np.pi)/m
t = (np.arange(n)*2*np.pi)/n
 
#True field
Xt = np.outer(np.sin(x), np.sin(t)) + \
 np.outer(np.sin(2.1*x), np.sin(2.1*t)) + \
 np.outer(np.sin(3.1*x), np.sin(3.1*t)) + \
 np.outer(np.tanh(x), np.cos(t)) + \
 np.outer(np.tanh(2*x), np.cos(2.1*t)) + \
 np.outer(np.tanh(4*x), np.cos(0.1*t)) + \
 np.outer(np.tanh(2.4*x), np.cos(1.1*t)) + \
 np.tanh(np.add.outer(x, t)) + \
 np.tanh(np.add.outer(x, 2*t))

fig, axes = subplots(1,3)
fig.set_size_inches(15,5)
im = axes[0].imshow(Xt,cmap=cm.Wistia,vmin=0,vmax=7)
axes[0].set_title('Synthetic Function')
 
#Noise field
RAND = np.random.rand(Xt.shape[0],Xt.shape[1])
R = RAND * NSratio * Xt

# #True field + Noise field
Xp = Xt + R
axes[1].imshow(Xp,cmap=cm.Wistia,vmin=0,vmax=7)
axes[1].set_title('Synthetic Function w/ Noise')
 
#Observed field with gaps
gaps = sorted(random.sample(xrange(Xp.size), int(frac_gaps*Xp.size)))
XG = np.copy(Xp)
XG = np.reshape(XG,Xp.shape[0]*Xp.shape[1])
XG[gaps] = np.nan
XG = np.reshape(XG,(Xp.shape[0],Xp.shape[1]))
axes[2].imshow(np.reshape(XG,(Xp.shape[0],Xp.shape[1])),cmap=cm.Wistia,vmin=0,vmax=7)
axes[2].set_title('Noise + Gaps (50% gaps, 10% noise)')

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.83, 0.33, 0.01, 0.35])
fig.colorbar(im, cax=cbar_ax)
```

```python
####################################################################################################
# Functions
####################################################################################################
X = np.copy(XG)
##### Iterative eof 
# def naneof_BR2003(X):
# Remove nanmean of data
dat_mean = np.nanmean(X)
X[~np.isnan(X)] = X[~np.isnan(X)]-dat_mean

# Subset of data for validation
X0 = np.copy(X)
idok_r,idok_c = np.nonzero(~np.isnan(X)) 
id_val_r = idok_r[np.ceil(np.random.uniform(size=(np.floor(idok_r.shape[0]/50)))* \
                          idok_r.shape[0]).astype(int)] # validation subset (rows)
id_val_c = idok_c[np.ceil(np.random.uniform(size=(np.floor(idok_c.shape[0]/50)))* \
                          idok_c.shape[0]).astype(int)] # validation subset (columns)

X_val = X[id_val_r,id_val_c]
mx_val = np.nansum(np.nansum(X_val**2))

# remove validation subset from the data
X[id_val_r,id_val_c] = np.nan;
# Update index of non-nans
idok_r,idok_c = np.nonzero(~np.isnan(X))
# replace NaNs by zeros
X[np.isnan(X)] = 0
```

```python
Nit = 100
tol = 1e-8
err = np.empty(min(X.shape[1],20))*np.nan
# find out how many eigenfunctions to retain...
for Ne in np.arange(1,min(X.shape[1],20)):
    X1 = np.copy(X)
    for k in np.arange(2,Nit):
        # compute SVD (equivalent of economy method in matlab)
        U,D,V = np.linalg.svd(X1,full_matrices=False)

        # truncate and estimate "interpolated" D
        N = Ne
        Ut = U[:,0:N]
        Dt = D[0:N]
        Vt = V[0:N,:]
        Xa = np.dot(np.dot(Ut,np.diag(Dt)),Vt)
        Xa[idok_r,idok_c] = X[idok_r,idok_c] # restore real data
        X2 = np.copy(Xa)

        # termination criterium?
        dx=np.nansum((X2-X1)**2,axis=1)
        dx=dx[:,np.newaxis]
        mx=np.nansum(X2**2,axis=1)
        mx=mx[:,np.newaxis]
        dxex = np.linalg.lstsq(dx.T,mx.T)
        dxex = dxex[-1]
        if dxex < tol:
            print('Converged in '+str(k-1)+' iterations to the tolerance of '+str(tol))
            break
        X1 = np.copy(X2)

    # error?
    Xa = np.dot(np.dot(Ut,np.diag(Dt)),Vt)
    dx_val = np.nansum(np.nansum((Xa[id_val_r,id_val_c]-X_val)**2))
    err[Ne] = dx_val/mx_val
```

```python
## Loop through again using the optimal number of eigenvalues only
Nopt = np.flatnonzero(err == np.nanmin(err))

X1 = np.copy(X0);
# Subset of data for validation
idok_r,idok_c = np.nonzero(~np.isnan(X1)) 
X1[np.isnan(X1)] = 0
for k in np.arange(2,Nit):
    # compute SVD (equivalent of economy method in matlab)
    U,D,V = np.linalg.svd(X1,full_matrices=False)#     N = Nopt;
    # truncate and estimate "interpolated" D
    N = Nopt
    Ut = U[:,0:N]
    Dt = D[0:N]
    Vt = V[0:N,:]
    Xa = np.dot(np.dot(Ut,np.diag(Dt)),Vt)
    Xa[idok_r,idok_c] = X[idok_r,idok_c] # restore real data
    X2 = np.copy(Xa)
    
    # termination criterium?
    dx=np.nansum((X2-X1)**2,axis=1)
    dx=dx[:,np.newaxis]
    mx=np.nansum(X2**2,axis=1)
    mx=mx[:,np.newaxis]
    dxex = np.linalg.lstsq(dx.T,mx.T)
    dxex = dxex[-1]
    if dxex < tol:
        print('Converged in '+str(k-1)+' iterations to the tolerance of '+str(tol))
        break
    X1 = np.copy(X2)
```

```python
# units in B
B = np.dot(U,D)  #the temporal modes
amps = V # the spatial modes
#the variance of the original temperature series
orig_var = np.nanmean(XG**2) 
#divide by the product of the original matrix size and by the variance
variance = D**2/orig_var*100/(X0.shape[0]*X0.shape[1]) 
```
