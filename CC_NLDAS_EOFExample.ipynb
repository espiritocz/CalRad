```python
##### Header #####
# Imports NLDAS SWdwn (in netcdf) for Jan 2004-2009 (see CC_NLDAS_Reformat for details)
# Performs EOF/PC analysis, creates plots

# Compile and write NLDAS SWdwn over the CA study area for Jan 2005 (example)
# netcdf/numpy
from netCDF4 import Dataset
from netCDF4 import num2date, date2num
import numpy as np
from datetime import datetime, timedelta

# OS interaction
import sys
import os
```

```python
## Directory of .nc files
wd = '/Users/karllapo/gdrive/SnowHydrology/proj/CloudClimatology/data/NLDAS'
pd = '/Users/karllapo/gdrive/SnowHydrology/proj/CloudClimatology/Graphics'
os.chdir(wd)
f = Dataset('NLDAS_SWdwn_Jan.nc','r')
```

```python
## Read and assign data
SWdwn = f.variables['DSWRF_110_SFC']
t = f.variables['time']

# I messed up making this file (didn't assign [lat/lon]_nc_out correctly. This is my work around 
#     -- upload new version of CC_NLDAS_Reformat.py
#     -- set up paths
#     -- remake NLDAS_SWdwn_Compile.nc
#     -- re-run this analysis without this temporary workaround
#lat = f.variables['latitude']
#lon = f.variables['longitude']
f_ll = Dataset('NLDAS_latlon_workaround.nc','r')
lat = f_ll.variables['latitude']
lon = f_ll.variables['longitude']
nlat = lat.size
nlon = lon.size

# Convert time to PST
times = num2date(t[:],t.units,t.calendar)-timedelta(hours=8)
t_num = date2num(times,t.units)

# Sort according to date
ind = times.argsort()
times = sorted(times)
SWdwn = SWdwn[ind]
t_num = t_num[ind]
```

```python
## Daily averages
# number date @ beginning of each day
t_days = np.unique(np.floor(t_num/24)*24)

# mean over each day
SWdwn_daily = np.empty((t_days.size,nlat,nlon))
for day in range(t_days.size):
    SWdwn_daily[day,:,:] = np.mean(SWdwn[(t_num[:] > t_days[day]) & (t_num[:] < t_days[day]+1*24),:,:],axis=0)
```

```python
## Monthly averages
# number date @ beginning of each month
hours_per_jan = 31*24
t_months = np.unique(np.floor(t_num/hours_per_jan)*hours_per_jan)

# mean over each day
SWdwn_monthly = np.empty((t_months.size,nlat,nlon))
for month in range(t_months.size):
    SWdwn_monthly[month,:,:] = np.mean(SWdwn[(t_num[:] > t_months[month]) & (t_num[:] < t_months[month]+hours_per_jan),:,:],axis=0)
```

```python
## Domain averages

# mean over each day
SWdwn_domainbar = np.empty((t_days.size,nlat,nlon))
for day in range(t_days.size):
    SWdwn_domainbar[day,:,:] = np.mean(SWdwn_daily[day])
```

```python
###### EOF ######
# Borrowing heavily from example file for eofs: http://ajdawson.github.io/eofs/examples/nao_standard.html
from eofs.standard import Eof

# EOF with spatially varying average over time
SWdwn_mean = np.mean(SWdwn,axis=0)
SWdwn_eof = SWdwn_daily-SWdwn_mean
coslat = np.cos(np.deg2rad(lat))
wgts = np.sqrt(coslat)[..., np.newaxis]
solver = Eof(SWdwn_eof, weights=wgts)
# Retrieve the leading EOFs, expressed as the covariance between the leading PC
# time series and the input SWdwn anomalies at each grid point.
lead_eof_s = solver.eofsAsCorrelation(neofs=3)
lead_pcs_s = solver.pcs(npcs=3, pcscaling=1)
eof_frac_var_s = solver.varianceFraction(neigs=10)

# EOF with temporally varying average over space
SWdwn_mean = np.mean(SWdwn,axis=0)
SWdwn_eof = SWdwn_daily-SWdwn_domainbar
coslat = np.cos(np.deg2rad(lat))
wgts = np.sqrt(coslat)[..., np.newaxis]
solver = Eof(SWdwn_eof, weights=wgts)
# Retrieve the leading EOFs, expressed as the covariance between the leading PC
# time series and the input SWdwn anomalies at each grid point.
lead_eof_t = solver.eofsAsCorrelation(neofs=3)
lead_pcs_t = solver.pcs(npcs=3, pcscaling=1)
# Fraction of variance explained
eof_frac_var_t = solver.varianceFraction(neigs=10)
```

```python
##### EOF - pretreatment: Elevation Angle ######
# Above EOFs are mostly an expression of the seasonal cycle or red noise. Instead, pre-treat to transmissivity

# Import solar geometry code
import solargeo

# Datetime object (UTC)
NLDAS_datetime_H = num2date(t[:],t.units,t.calendar)
ind = NLDAS_datetime_H.argsort()
NLDAS_datetime_H = np.array(sorted(NLDAS_datetime_H))

# Solar Zenith Angle and Azimuth Angle for each timestep, by month
lat_size = np.shape(lat)
lon_size = np.shape(lon)
t_size = np.shape(NLDAS_datetime_H)

# Problem here -- if statement checks, when passing in a single time, often fail since the 'inds' arguments pass out an
# empty value. Can be fixed by adding a check "if not inds:" -- need to convert both the .ipynb and the .pyx, plus rebuild

# Other problem: I can't get the NLDAS_datetime_h object to return the time series of 'year','jday', or 'hour' since it is a
# numpy array. Need to do some cleaning up of this code before it goes to primetime.

# Both problems fixed below!
EL = np.zeros((lon_size[0],lat_size[0],t_size[0]))
yyyy = np.zeros(t_size[0])
jday = np.zeros(t_size[0])
hh = np.zeros(t_size[0])

for ind,time_step in enumerate(NLDAS_datetime_H):
    yyyy[ind] = np.array([time_step.year])
    jday[ind] = np.array([time_step.timetuple().tm_yday])
    hh[ind] = np.array([time_step.hour+time_step.minute/60.])

# mesh lat, lon, and time variables
lat_m,lon_m,yyyy_m = np.meshgrid(lat,lon,yyyy)
jday_m = np.meshgrid(lat,lon,jday)[2]
hh_m = np.meshgrid(lat,lon,hh)[2]

# Elevation Angle (inst.) @ all times and locations
EL = solargeo.SUNAE(yyyy_m,jday_m,hh_m,lat_m,lon_m,refraction_flag=0)[0]
print(EL.shape)
```

```python
##### EOF - pretreatment: Transmissivity ######

# SWdwn and Elevation Angle dimensions are mixed up -- reshape!
# SWdwn_re = np.empty((44,32,4464))
# for dim0,dim1 in zip(np.arange(0,44),np.arange(0,32)):
#     SWdwn_re[dim0,dim1,:] = SWdwn[:,dim1,dim0]

# Now Tau is mis-shapen for the EOF -> reshape back
EL_re = np.empty((4464,32,44))
for dt in np.arange(0,4464):
    EL_re[dt,:,:] = EL[:,:,dt].conj().transpose()
# for dim1,dim2 in zip(np.arange(0,32),np.arange(0,44)):
#     EL_re[:,dim1,dim2] = EL[dim2,dim1,:]

# Find transmissivity
Tau = SWdwn/(1365*np.sin(EL_re*np.pi/180.))
# Enforce physicality, remove marginal cases where a refraction
# correction will create inaccurate transmissivities.
Tau[Tau==0]=np.nan
Tau[EL_re < 9]=np.nan

## Notes:
# This transmissivity calculation still creates spurious diurnal fluctuations.
# For now, I'm allowing this. Future use of this code block may want to be aware of the issue.
# It is explicitly not related to time stamp issues--already checked.

## Code for spot checking
# fig, ax = subplots(2,1)
# fig.set_size_inches(15,5)
# ind = np.arange(0,4400)
# ax[0].plot(1365*np.sin(EL[0,0,ind]*np.pi/180))
# ax[0].plot(SWdwn_re[0,0,ind])
# ax[1].plot(Tau[15,15,ind])

## Problem:
# Transmissivity doesn't appear to have been calculated correctly for grid points other than 0,0. 
# Below code is for exploring that.
## Solution:
# I was somehow shredding my data set by reshaping it--time was occuring across the spatial domain.
# This is now fixed by using the 'conj().transpose()' command to transpose the matrix to the right shape.
# Looking at xray--seems like it would automatically handle a lot of these sorts of annoying things.
# fig, ax = subplots(1,2)
# fig.set_size_inches(15,5)
# ind = np.arange(0,200)
# for nlat in np.arange(0,32):
#     ax.plot(NLDAS_datetime_H[ind],SWdwn[ind,nlat,0],label=str(nlat))
# ax.legend()
# ax[0].imshow(Tau[20,:,:])
# ax[1].imshow(SWdwn[20,:,:])
```
