```python
##### Header #####
# Imports NLDAS SWdwn (in netcdf) for Jan 2004-2009 (see CC_NLDAS_Reformat for details)
# Performs EOF/PC analysis, creates plots

# Compile and write NLDAS SWdwn over the CA study area for Jan 2005 (example)
# netcdf/numpy
from netCDF4 import Dataset
from netCDF4 import num2date, date2num
import numpy as np
from datetime import datetime, timedelta

# OS interaction
import sys
import os
```

```python
## Directory of .nc files
wd = '/Users/karllapo/Dropbox/SnowHydrology/proj/CloudClimatology'
pd = '/Users/karllapo/Dropbox/SnowHydrology/proj/CloudClimatology/Graphics'
os.chdir(wd)
f = Dataset('NLDAS_SWdwn_Compiled.nc','r')
```

```python
## Read and assign data
SWdwn = f.variables['DSWRF_110_SFC']
t = f.variables['time']

# I messed up making this file (didn't assign [lat/lon]_nc_out correctly. This is my work around 
#     -- upload new version of CC_NLDAS_Reformat.py
#     -- set up paths
#     -- remake NLDAS_SWdwn_Compile.nc
#     -- re-run this analysis without this temporary workaround
#lat = f.variables['latitude']
#lon = f.variables['longitude']
f_ll = Dataset('NLDAS_latlon_workaround.nc','r')
lat = f_ll.variables['latitude']
lon = f_ll.variables['longitude']
nlat = lat.size
nlon = lon.size

# Convert time to PST
times = num2date(t[:],t.units,t.calendar)-timedelta(hours=8)
t_num = date2num(times,t.units)

# Sort according to date
ind = times.argsort()
times = sorted(times)
SWdwn = SWdwn[ind]
t_num = t_num[ind]
```

```python
## Daily averages
# number date @ beginning of each day
t_days = np.unique(np.floor(t_num/24)*24)

# mean over each day
SWdwn_daily = np.empty((t_days.size,nlat,nlon))
for day in range(t_days.size):
    SWdwn_daily[day,:,:] = np.mean(SWdwn[(t_num[:] > t_days[day]) & (t_num[:] < t_days[day]+1*24),:,:],axis=0)
```

```python
## Monthly averages
# number date @ beginning of each month
hours_per_jan = 31*24
t_months = np.unique(np.floor(t_num/hours_per_jan)*hours_per_jan)

# mean over each day
SWdwn_monthly = np.empty((t_months.size,nlat,nlon))
for month in range(t_months.size):
    SWdwn_monthly[month,:,:] = np.mean(SWdwn[(t_num[:] > t_months[month]) & (t_num[:] < t_months[month]+hours_per_jan),:,:],axis=0)
```

```python
## Domain averages

# mean over each day
SWdwn_domainbar = np.empty((t_days.size,nlat,nlon))
for day in range(t_days.size):
    SWdwn_domainbar[day,:,:] = np.mean(SWdwn_daily[day])
```

```python
###### EOF ######
# Borrowing heavily from example file for eofs: http://ajdawson.github.io/eofs/examples/nao_standard.html
from eofs.standard import Eof

# EOF with spatially varying average over time
SWdwn_mean = np.mean(SWdwn,axis=0)
SWdwn_eof = SWdwn_daily-SWdwn_mean
coslat = np.cos(np.deg2rad(lat))
wgts = np.sqrt(coslat)[..., np.newaxis]
solver = Eof(SWdwn_eof, weights=wgts)
# Retrieve the leading EOFs, expressed as the covariance between the leading PC
# time series and the input SWdwn anomalies at each grid point.
lead_eof_s = solver.eofsAsCorrelation(neofs=3)
lead_pcs_s = solver.pcs(npcs=3, pcscaling=1)
eof_frac_var_s = solver.varianceFraction(neigs=10)

# EOF with temporally varying average over space
SWdwn_mean = np.mean(SWdwn,axis=0)
SWdwn_eof = SWdwn_daily-SWdwn_domainbar
coslat = np.cos(np.deg2rad(lat))
wgts = np.sqrt(coslat)[..., np.newaxis]
solver = Eof(SWdwn_eof, weights=wgts)
# Retrieve the leading EOFs, expressed as the covariance between the leading PC
# time series and the input SWdwn anomalies at each grid point.
lead_eof_t = solver.eofsAsCorrelation(neofs=3)
lead_pcs_t = solver.pcs(npcs=3, pcscaling=1)
# Fraction of variance explained
eof_frac_var_t = solver.varianceFraction(neigs=10)
```

```python
###### Graphing ######

# import subplots function for plotting
from matplotlib.pyplot import subplots
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.basemap import Basemap

# must insert this statement to render the plots within the notebook
# this is specific to the ipython notebook
%matplotlib inline
```

```python
## Plot monthly averages
# I'm stealing a lot of code/guidance from http://matplotlib.org/basemap/users/examples.html
fig, ax = subplots(1,6)
fig.set_size_inches(15,2.5)
years = range(1,12,2)

for (num,year) in enumerate(years):
    m = Basemap(llcrnrlon=lon[0],llcrnrlat=lat[0],urcrnrlon=lon[-1],urcrnrlat=lat[-1],\
            rsphere=(6378137.00,6356752.3142),\
            resolution='l',area_thresh=1000.,projection='lcc',\
            lat_1=lat[16],lon_0=lon[20],ax=ax[num])
    im = m.imshow(SWdwn_monthly[year,:,:],cmap=cm.gnuplot2)
    im.set_clim(vmin = 100, vmax = 150)
    # draw coastlines and political boundaries.
    m.drawcoastlines()
    m.drawcountries()
    m.drawstates()
    m.drawcounties()
    # Meridians and Parallels
    if (num == 0):
        m.drawparallels(np.arange(np.round(lat[0]), np.round(lat[-1]), .5),labels=[1,0,0,0])
    else:
        m.drawparallels(np.arange(np.round(lat[0]), np.round(lat[-1]), .5))        

    m.drawmeridians(np.arange(np.round(lon[0]), np.round(lon[-1]), 5),labels=[0,0,0,1])   
    
    if (num == 5):
        cb = m.colorbar(im,"right", size="5%", pad='2%')
        cb.set_label('Wm-2', fontsize=12)

plt.tight_layout()

os.chdir(pd)
fname = 'NLDAS_Jan_2004_2009.MonthlyMean.pdf'
fig.savefig(fname)
```

```python
## EOF w/ space constant mean removed
fig, axs = subplots(1,3)
fig.set_size_inches(15,5)

for eof_num in range(0,3):
    m = Basemap(llcrnrlon=lon[0],llcrnrlat=lat[0],urcrnrlon=lon[-1],urcrnrlat=lat[-1],\
        rsphere=(6378137.00,6356752.3142),\
        resolution='l',area_thresh=1000.,projection='lcc',\
        lat_1=lat[16],lon_0=lon[20],ax=axs[eof_num])
    # Image of leading eofs
    im = m.imshow(lead_eof_s[eof_num,:,:],cmap=cm.RdBu_r)
    im.set_clim(vmin = -1, vmax = 1)
    # draw coastlines and political boundaries.
    m.drawcoastlines()
    m.drawcountries()
    m.drawstates()
    m.drawcounties()
    # Meridians and Parallels
    if eof_num == 0:
        m.drawparallels(np.arange(np.round(lat[0]), np.round(lat[-1]), .5),labels=[1,0,0,0]) 
    else:
        m.drawparallels(np.arange(np.round(lat[0]), np.round(lat[-1]), .5))         
    m.drawmeridians(np.arange(np.round(lon[0]), np.round(lon[-1]), 5),labels=[0,0,0,1])
    # Colorbar
    if eof_num == 2:
        cb = m.colorbar(im,"right", size="5%", pad='2%')
        cb.set_label('correlation coefficient', fontsize=12)

plt.tight_layout()

os.chdir(pd)
fname = 'NLDAS_Jan_2004_2009.n3EOF_SpConMean.pdf'
fig.savefig(fname)
```
