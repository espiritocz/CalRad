```python
#################
## Climatology ##
#################
def mon_clim(xray_df):
    # Monthly climatology values
    df_cl = xray_df.groupby('time.month').mean('time')
    # Monthly anomaly maps
    df_anom = xray_df.groupby('time.month')-df_cl

    # Return values
    return(df_cl,df_anom)

monthly_cl = {}
monthly_anom = {}
monthly_cl['SYN'],monthly_anom['SYN'] = mon_clim(syn_monthly)
monthly_cl['NLDAS'],monthly_anom['NLDAS'] = mon_clim(nldas_monthly)

# # Monthly climatology values
# syn_monthly_cl = syn_monthly.groupby('time.month').mean('time')
# nldas_monthly_cl = nldas_monthly.groupby('time.month').mean('time')

# # Monthly anomaly maps
# syn_monthly_anom = syn_monthly.groupby('time.month')-syn_monthly_cl
# nldas_monthly_anom = nldas_monthly.groupby('time.month')-nldas_monthly_cl
print(monthly_cl['NLDAS'],monthly_anom['NLDAS'])
```
