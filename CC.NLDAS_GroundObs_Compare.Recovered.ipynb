```python
## NLDAS
os.chdir(proj_dir+NLDAS_dir)
f = Dataset('NLDAS_SWdwn_Jan.nc','r')

## Read and assign data
SWdwn = f.variables['DSWRF_110_SFC']
t = f.variables['time']

# I messed up making this file (didn't assign [lat/lon]_nc_out correctly. This is my work around 
#     -- upload new version of CC_NLDAS_Reformat.py
#     -- set up paths
#     -- remake NLDAS_SWdwn_Compile.nc
#     -- re-run this analysis without this temporary workaround
#lat = f.variables['latitude']
#lon = f.variables['longitude']
f_ll = Dataset('NLDAS_latlon_workaround.nc','r')
lat = f_ll.variables['latitude']
lon = f_ll.variables['longitude']
nlat = lat.size
nlon = lon.size

# Convert time to PST
times = num2date(t[:],t.units,t.calendar)-timedelta(hours=8)
t_num = date2num(times,t.units)

# Sort according to date
ind = times.argsort()
times = sorted(times)
SWdwn = SWdwn[ind]
t_num = t_num[ind]

## Daily Averages ##
# number date @ beginning of each day
t_days = np.unique(np.floor(t_num/24)*24)

# mean over each day
SWdwn_daily_NLDAS = np.empty((t_days.size,nlat,nlon))
for day in range(t_days.size):
    SWdwn_daily_NLDAS[day,:,:] = np.mean(SWdwn[(t_num[:] > t_days[day]) & (t_num[:] < t_days[day]+1*24),:,:],axis=0)
```

```python
## Ground Obs - SIO stations
os.chdir(proj_dir+ground_obs_dir)
content = os.listdir(os.getcwd())
num_files = len([name for name in os.listdir('.') if os.path.isfile(name)])

# Empty dictionary literal
tabl = {}
SWdwn_daily_GrObs = {}

# time zone variables
tz_pst = pytz.timezone('US/Pacific')

for files in content:
    # Only read QC formatted files
    if files[-10:] == 'Rad.QC.txt':
        
        sitename = files.split('.')[0]
        
        # Future: 
        #     1)write DAN, TES, and TUM with python readable dates
        #     2)skip Edwin's DAN, TES, & TUM and use mine instead
        if sitename == 'DAN' or sitename == 'TUM' or sitename == 'TES':
            continue
        
        na_value = ['   NaN']
        
        # Read SW data, convert from UTC to PST, and get SW that passes QC
        tabl[sitename] = pd.read_csv(files,sep= '\t', parse_dates=True, index_col=0, na_values=na_value)
        tabl[sitename].index = tabl[sitename].index.tz_localize(pytz.utc).tz_convert(tz_pst)
        tabl[sitename]['SWdwn_QC'] = tabl[sitename]['SWdwn_Wm^-2'].where(tabl[sitename]['QCFlag'] == 0)

        # New data frame w/ daily means
        SWdwn_daily_GrObs[sitename] = tabl[sitename]['SWdwn_Wm^-2'].resample('D', how='mean').to_frame(name='SWdwn_D')
        SWdwn_daily_GrObs[sitename]['SWdwn_D_QC'] = tabl[sitename]['SWdwn_QC'].resample('D', how='mean')
        SWdwn_daily_GrObs[sitename]['SWdwn_D_proc'] = tabl[sitename]['SWdwn_proc'].resample('D', how='mean')

## Read station information
stdat = pd.read_csv('SIO.stninfo.txt',sep= '\t', index_col=1) # Read the supporting information
```
